This network will attempt to predict the mpg for a car given its: num of cyclinders, horsepower, weight, and acceleration.



    2. cylinders:     multi-valued discrete
    3. displacement:  continuous
    4. horsepower:    continuous
    5. weight:        continuous
    6. acceleration:  continuous

```python
import numpy as np
import random as rand
import tensorflow as tf
import pandas as pd
import keras
import matplotlib
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import SimpleRNN
from keras.layers import Add as keras_add_layers
from keras.callbacks import History
np.set_printoptions(suppress=True)
```

```python
# Removes the provided character from a given string.
def remove_character(text,character):
    array = text.split(character)
    proxy_array = []
    for val in array:
        if val != character and val != '':
            proxy_array.append(val)
    return proxy_array
    
# Loads the training data.
def load_training_data():
    try:
        file = open('car_mpg.txt')
        file_text = file.read()
        file_lines = file_text.split('\n')
        inputs = []
        outputs = []
        for line in file_lines:
            if line.find('?') == -1:
                vals = remove_character(line,' ')
                inputs.append([float(vals[1]),float(vals[3]),float(vals[4])])
                outputs.append(float(vals[0]))
        return [np.array(inputs),np.array(outputs)]
    except:
        print('Training data file could not be found.')
        return None
```

```python
# Loading in training data.
training_data = load_training_data()
# Seperating data into input and output.
training_inputs = training_data[0]
training_outputs = training_data[1]
```

```python
# Create model
system_model = Sequential()
#system_model.add(SimpleRNN(2,activation="linear"))
system_model.add(Dense(1,activation="linear",input_dim=3))
#system_model.add(Dropout(0.12))
#system_model.add(Dense(2,activation="relu"))
#system_model.add(Dropout(0.12))
#system_model.add(Dense(1,activation="linear"))
system_model.compile(optimizer="adamax",loss="mean_absolute_percentage_error",metrics=['mae'])
```

```python
# Loading in training data.
training_data = load_training_data()
# Seperating data into input and output.
training_inputs = training_data[0][:330]
training_outputs = training_data[1][:330]

testing_inputs = training_data[0][330:]
testing_outputs = training_data[1][330:]

keras_history = system_model.fit(training_inputs,training_outputs,batch_size=32,epochs=10000,validation_split=0.2)
```

```python
total_difference = 0
for i in range(len(testing_inputs)):
    real = testing_outputs[i]
    predicted = system_model.predict(testing_inputs[i:i+1])[0][0]
    difference = abs(real - predicted) / ((real + predicted) / 2) 
    difference = 100 - (difference * 100)
    print(f'Real : {real} -- Predicted : {predicted} -- Difference : {difference}%')
    total_difference += difference

print(f'Average Similarity : {total_difference / len(training_inputs)}%')
```
